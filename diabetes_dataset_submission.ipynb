```python
import numpy as np
from sklearn.linear_model import LinearRegression
algo = LinearRegression()
file ='/Users/shaurryalakhanpal/diab.csv'
# train data 
data =np.loadtxt(file,delimiter = ',')
l = len(data[0])
x_train = data[:,0:l-1]
y_train = data[:,l-1]
algo.fit(x_train,y_train)
#import test data 
file = '/Users/shaurryalakhanpal/diab_test.csv'
x_test =np.loadtxt(file,delimiter = ',')
y_pred = algo.predict(x_test)
print(y_pred)
y_pred=[round(i,5) for i in y_pred]
y_pred = np.array(y_pred)
np.savetxt("predictions.csv", y_pred, delimiter=",")


```
